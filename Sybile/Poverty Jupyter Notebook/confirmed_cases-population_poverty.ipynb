```python
import pandas as pd
import os as csv
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

#path to files
path1 = '../Poverty Data/1_county_level_confirmed_cases.csv'
path2 = '../Poverty Data/Population_and_Poverty_-_Counties.csv'

#read in files
csvfile1 = pd.read_csv(path1)
csvfile2 = pd.read_csv(path2)

#Convert files to data frame
confirmed_cases = pd.DataFrame(csvfile1)
population_poverty = pd.DataFrame(csvfile2)
```

```python
#Display confirmed_cases
confirmed_cases
```

```python
#Display population_poverty
population_poverty
```

```python
#Add country column
population_poverty['country'] = 'US'
population_poverty
```

```python
# Create new column by splitting 'GEO_NAME' column and selecting the first word (county name)
population_poverty['county_name'] = population_poverty['GEO_NAME'].str.split(' ', 1, expand=True)[0]
population_poverty
```

```python
#Create new column by merging county_name, GEO_PARENT_NAME and country
population_poverty['county_name_long'] = population_poverty['county_name'] + ', ' + population_poverty['GEO_PARENT_NAME'] + ', ' + population_poverty['country']
population_poverty
```

```python
#Select the columns needed for analysis
population_poverty = population_poverty.loc[:, ['county_name_long', 'HOUSEBELOWPOVP_CALC']]
population_poverty
```

```python
#Merge both data frames though inner join on county_name_long
merged_data = pd.merge(confirmed_cases, population_poverty, on='county_name_long', how='inner')
merged_data.head()
```

```python
#Save Output file as .csv
merged_data.to_csv('../Poverty Data/confirmed_cases-population_poverty.csv', index=False)
```

```python
#Create Scatter plot to analyze data
plt.scatter(merged_data['HOUSEBELOWPOVP_CALC'], merged_data['confirmed_per_100000'], color='green', edgecolor='black')
plt.title("Confirmed Cases vs. Percentage of Household Below Income Level")
plt.xlabel("Household Below Income Level")
plt.ylabel("Confirmed cases per 100,000")

#Calculate Linear Regression Model
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(merged_data['HOUSEBELOWPOVP_CALC'], 
                                                           merged_data['confirmed_per_100000'])
regress_value = merged_data['HOUSEBELOWPOVP_CALC'] * slope + intercept
plt.plot(merged_data['HOUSEBELOWPOVP_CALC'], regress_value, 'r-')
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(30, 10000),fontsize=15,color="red")
plt.grid()
print(f'The p-value is: {pvalue}')
print(f'The r-value is: {rvalue}')

#Save chart as .png
plt.savefig("../Poverty Images/Confirmed Cases vs. Percentage of Household Below Income Level.png")
```
